***10-K Annual Reports***

Publicly- traded companies are required by the U.S Securities and Exchange Commission (SEC) to annually file a comprehensive report about their financial performances, named "10-k". With this report, the SEC requires to keep investors well- informed about companies' financial condition, so they have fundamental information about them, for making knowledgeable investment decisions. 

The report is lengthy and detailed in order to give a clearer picture of a company's activities and kind of risks it faces. The report includes five different sections: (1) Business (an overview of the company’s main operations), (2) Risk factors (any and all risks the company faces or may face in the future), (3) Selected financial data (Selected financial data), (4) Management’s discussion and analysis (of financial condition and results of operations), (5) Financial statements and supplementary data (includes the company’s audited financial statements including the income statement, balance sheets and statement of cash flows). 

In this project, the focus will be on section 1A. Risk Factors, in which firms are legally obliged to discuss "the most significant factors that make the company speculative or risky". 

The investigated companies are the ones who have experienced "Cyber Incidents", according to the PRC data base (see below) . We extract the "Risk Factors" section out of the companies' 10-k Annual Reports of the previous year to the incident. ##    1.2 Extracting section "1.A Risk Factors"

In this part we validate and manipulate the text in the files, in order to extract section "1A. Risk Factors". 

***Check: if a word  exists in a line***

```python
def isWordInLine(word, line):
    if line.find(word) != -1:
        return True
    return False
```

***Check: if the string names line contains any of the starts key words***

```python
def isRiskparagraphLine(line):
    index = 0
    for word in startKeyWords:
        if isWordInLine(word, line):
            if startKeyWordCounters[index] == 0:
                startKeyWordCounters[index] = 1
                break
            elif startKeyWordCounters[index] == 1:
                return True
        index += 1
    return False
```

***Check: if a line contains any key words that indicate reaching the end of a the section***

```python
def isRiskparagraphEnded(line):
    global createdFilesCounter
    for word in endKeyWords:
        if isWordInLine(word, line):
            createdFilesCounter += 1
            return True
    return False
```

## 2.2 Normalization:

***Check: if a char is valid***

```python
def isValidChar(c):
    if c.isdigit() or c.isalpha() or c =='\n' or c =='' or c==' 'or '\t': 
        return True
    return False
```

***Update: lower case all words***

```python
def getWordNormalized(word):
    normalizedWord = ''
    for c in word:
        if isValidChar(c):
            normalizedWord += c
    return normalizedWord.lower()
```

***Normalize a file buffer***

```python
def getBufferNormalized(buffer):
    normalizedBuffer = []
    for word in buffer:
        normalizedWord = getWordNormalized(word)
        normalizedBuffer.append(normalizedWord)
    return normalizedBuffer
```

***Why Lematizing and not Stemming?***

While pre-processing the text, we explored the usage of both Lemmatization and Stemming. Then we used both method for finding the intersecting set of words. When using Stemming, we achieved finding a few more words of the intersection. On the other hand, we found out that this usage takes words out of their context, a fact that can make our words set biased. After investigating some literature, we have decided to use Lemmatization, as this method uses the context in which the word is being used, instead of its stem (which is used in Stemming). This resulted in reduced data, while still keeping its context.

```python
def getFileFreqSeries(fileBuffer):
    words = fileBuffer.split()
    FreqDistList = FreqDist(words)
    percentage = 0.65
    size = int(len(words)* (1 - percentage))
    FreqDistList = FreqDistList.most_common(size)
    return FreqDistList
```

## 3.2 Intersection

***Intersect the collection's sets (can be done in different methods)***

```python
def getIntersectionBy(IntersectionType = ''):
    # import these modules
    ps = PorterStemmer()
    wnl = WordNetLemmatizer()
    
    coupleList = allFilesFreqTable[0]
    curIntersection = set()
    for item in curList:
        if IntersectionType == '':
            curIntersection.add(item[0]) 
        elif IntersectionType == 'ps':
            curIntersection.add(ps.stem(item[0]))
        elif IntersectionType == 'wnl':
            curIntersection.add(wnl.lemmatize(item[0]))

    i = 1

    for item in range (len(allFilesFreqTable) - 1):
        coupleList = allFilesFreqTable[i] # list of ordered couple of (word, frequency)    
        curSet = set()    
        for item in coupleList:

            if IntersectionType == '':
                curSet.add(item[0]) 
            elif IntersectionType == 'ps':
                curSet.add(ps.stem(item[0]))
            elif IntersectionType == 'wnl':
                curSet.add(wnl.lemmatize(item[0]))

        curIntersection = curIntersection.intersection(curSet)
        i += 1

    return curIntersection
```

***Make a two d list out of a series***

```python
def seriesToTwoDList(series):
    aDict = series.to_dict()
    twoDList = []
    for key, value in aDict.items():
        twoDList.append([round(value, 5), key])
    return twoDList
```

***Make a diagram of a series (scatter/histogram)***

```python
def seriesToDiagram(series, diagramType = 's'):
    seriesTwoDList = seriesToTwoDList(series)
    df = pd.DataFrame(seriesTwoDList, columns=['average', 'word'])

    
    #this if else is ready for much more diagrams
    if diagramType == 's':
        return df.plot.scatter(x='word',y='average',figsize=(10,5), s=100,c='DarkBlue')
    if diagramType == 'h':
        aDict = series.to_dict()
        plt.subplots(figsize=(10,5))
        return plt.bar(aDict.keys(), aDict.values(), 0.25, color='DarkBlue')
    else: #diagramType == somthing else
        return None
    
```

#### Build the plot

```python
def getMeanStdPlot():
    # Build the plot
    fig, ax = plt.subplots()
    ax.bar(x_pos, means,yerr=error,align='center',alpha=0.5,ecolor='black',capsize=10)
    ax.set_ylabel('mean + std scale')
    ax.set_xticks(x_pos)
    ax.set_xticklabels(labels)
    ax.set_title('mean and std plot for words probabilities')
    ax.yaxis.grid(True)

    # Save the figure and show
    plt.tight_layout()
    plt.savefig('bar_plot_with_error_bars.png')
    plt.show()
```
